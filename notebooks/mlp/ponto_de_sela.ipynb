```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# para figuras interativas usar 'notebook' ao inves de 'inline'
%matplotlib notebook

# FunÃ§Ã£o que possui um ponto de sela em (0, 0)
def saddle_point(x, y):
    return x**2 - y**2

# CriaÃ§Ã£o de pontos no espaÃ§o 3D
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = saddle_point(X, Y)

# ConfiguraÃ§Ã£o da figura 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plotagem da superfÃ­cie
ax.plot_surface(X, Y, Z, cmap='viridis')

# Ponto de sela em (0, 0)
saddle_point_value = saddle_point(0, 0)
#ax.scatter(0, 0, saddle_point_value, color='red', s=100, label='Ponto de Sela')


# ConfiguraÃ§Ã£o dos rÃ³tulos
ax.set_xlabel('$w_0$', fontsize=14)
ax.set_ylabel('$w_1$', fontsize=14)
ax.set_zlabel('$J(w)$', fontsize=14)
#ax.set_title('SuperfÃ­cie com Ponto de Sela')



# Adiciona a legenda
#ax.legend()

# ExibiÃ§Ã£o do grÃ¡fico
plt.show()
```
