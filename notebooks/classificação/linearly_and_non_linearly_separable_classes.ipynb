```python
# Import all necessary libraries.
import numpy as np
import random
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons
from sklearn.datasets import make_blobs
```

```python
# Define a seed.
seed = 42
# Reset the PN sequence generator.
np.random.seed(seed)
random.seed(seed)
```

```python
# Number of examples.
N = 1000
```

```python
# Create a 2-class dataset for classification.
centers = [[-3.0, 0], [3.0, 2.0]]
X, y = make_blobs(n_samples=N, centers=centers, cluster_std=0.7, random_state=seed)

x = np.linspace(-5,5,1000)
g = 1 - x

# Plot the classes.
idx0 = np.argwhere(y == 0)
idx1 = np.argwhere(y == 1)

fig = plt.figure()
plt.plot(X[idx1,0], X[idx1,1], 'rx', label='Class 0')
plt.plot(X[idx0,0], X[idx0,1], '.', label='Class 1')
plt.plot(x,g,'--k')
plt.xlabel('$x_1$', fontsize=14)
plt.ylabel('$x_2$', fontsize=14)
plt.ylim(-2,5)
plt.legend()
plt.grid()
plt.savefig(fname='linearly_separable.png', dpi=600)
plt.show()
```

```python
# Create a 2-class dataset for classification.
X, y = make_moons(n_samples=N, noise=.05, random_state=seed)

# Plot the two circles.
idx0 = np.argwhere(y == 0)
idx1 = np.argwhere(y == 1)

fig = plt.figure()
plt.plot(X[idx0,0], X[idx0,1], '.', markersize=10, label='Class 0')
plt.plot(X[idx1,0], X[idx1,1], 'r^', markersize=10, label='Class 1')
plt.xlabel('$x_1$', fontsize=14)
plt.ylabel('$x_2$', fontsize=14)
plt.legend()
plt.grid()
plt.savefig(fname='nonlinearly_separable.png', dpi=600)
plt.show()
```
