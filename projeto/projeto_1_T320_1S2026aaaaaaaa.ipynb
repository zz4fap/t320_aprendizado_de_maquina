# 2) **PROJETO SOBRE CLASSIFICAĂĂO APLICADA ĂS TELECOMUNICAĂĂES.**

**CONTEXTO: VocĂȘ Ă© um engenheiro de dados e foi contratado para desenvolver um sistema inteligente de demodulaĂ§ĂŁo de sinais modulados em 8-PSK utilizando classificadores de Machine Learning. Neste projeto, vocĂȘ irĂĄ gerar a constelaĂ§ĂŁo 8-PSK, simular o canal AWGN e treinar um classificador para recuperar os sĂ­mbolos transmitidos a partir das componentes I e Q.**




1. Execute a cĂ©lula abaixo e analise o resultado. A figura mostra sĂ­mbolos ruidosos da modulaĂ§ĂŁo 8-PSK, onde cada um dos 08 possĂ­veis sĂ­mbolos Ă© considerado como uma classe diferente.

**DICAS:**

+ Note que na cĂ©lula de cĂłdigo abaixo, o conjunto de dados jĂĄ estĂĄ dividido em conjuntos de treinamento e validaĂ§ĂŁo.
+ Veja as referĂȘncias abaixo.


#### **ReferĂȘncias:**

[1] '8-PSK', https://www.researchgate.net/publication/3158035_8-PSK_trellis_codes_for_a_Rayleigh_channel/link/0a85e5346717b0d657000000/download?_tp=eyJjb250ZXh0Ijp7ImZpcnN0UGFnZSI6InB1YmxpY2F0aW9uIiwicGFnZSI6InB1YmxpY2F0aW9uIn19

[2] 'Phase-shift_keying', https://en.wikipedia.org/wiki/Phase-shift_keying

````python
# Import all necessary modules.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import seaborn as sns
from scipy.special import erfc
import time
from scipy.spatial import Voronoi, voronoi_plot_2d

# Reset PN sequence generator.
seed = 42
np.random.seed(seed)

# FunĂ§ĂŁo de modulaĂ§ĂŁo 8-PSK.
def modulate8PSK(bits):
    """
    Modula bits (0 a 7) em sĂ­mbolos 8-PSK com energia normalizada para 1.
    """
    # Ăngulos para 8-PSK: 0Â°, 45Â°, 90Â°, 135Â°, 180Â°, 225Â°, 270Â°, 315Â°
    angles = np.array([0, 45, 90, 135, 180, 225, 270, 315]) * np.pi / 180.0

    # SĂ­mbolos complexos (antes da normalizaĂ§ĂŁo)
    ip = np.exp(1j * angles[bits])

    # NormalizaĂ§ĂŁo de energia para 1 (dividir por sqrt(2) nĂŁo Ă© necessĂĄrio aqui)
    symbol = ip / np.sqrt(1.0)   # |s| = 1 para todos os sĂ­mbolos
    return symbol

# ParĂąmetros.
N = 10000                      # NĂșmero de sĂ­mbolos
numOfClasses = 8               # SĂŁo 8 classes (8-PSK)
EsN0dB = 13                    # SNR em dB
EsN0Lin = 10.0 ** (-(EsN0dB / 10.0))   # SNR linear

# GeraĂ§ĂŁo dos sĂ­mbolos e canal AWGN (abordagem vetorizada).
# Gera os bits originais para os sĂ­mbolos 8-PSK.
bit_8psk = np.random.randint(0, numOfClasses, N)

# Modula os bits em sĂ­mbolos 8-PSK.
symbols = modulate8PSK(bit_8psk)

# Adiciona ruĂ­do AWGN aos sĂ­mbolos.
noise = np.sqrt(EsN0Lin / 2.0) * (np.random.randn(N) + 1j * np.random.randn(N))
y = symbols + noise

# CriaĂ§ĂŁo da matriz de atributos (features).
X = np.c_[np.real(y), np.imag(y)]   # Features: parte real + parte imaginĂĄria

# Esplitagem em treino e teste.
X_train, X_test, y_train, y_test = train_test_split(
    X, bit_8psk, test_size=0.3, random_state=seed, stratify=bit_8psk
)

# Plotagem das constelaĂ§Ă”es 8-PSK.
plt.figure(figsize=(10, 8))
colors = ['black', 'red', 'blue', 'green', 'orange', 'purple', 'brown', 'magenta']
labels = [f'Symbol {i}' for i in range(8)]

for i in range(8):
    idx = np.argwhere(bit_8psk == i)
    plt.plot(np.real(y[idx.ravel()]), np.imag(y[idx.ravel()]), '.',
             color=colors[i], label=labels[i], markersize=8)

plt.grid(True, alpha=0.3)
plt.xlabel('In-Phase (Real)', fontsize=12)
plt.ylabel('Quadrature (Imag)', fontsize=12)
plt.title(f'ConstelaĂ§ĂŁo 8-PSK com AWGN (Es/N0 = {EsN0dB} dB)', fontsize=14)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.axis('equal')
plt.tight_layout()
plt.show()

# InformaĂ§Ă”es gerais.
print(f"   GeraĂ§ĂŁo de dados 8-PSK concluĂ­da.")
print(f"   Total de sĂ­mbolos: {N}")
print(f"   NĂșmero de classes: {numOfClasses}")
print(f"   SNR (Es/N0): {EsN0dB} dB")
print(f"   Shape de X: {X.shape}")
print(f"   Shape de y: {bit_8psk.shape}")
print(f"   Classes Ășnicas: {np.unique(bit_8psk)}")
````

Portanto, execute a cĂ©lula de cĂłdigo abaixo e analise o resultado obtido.

**DICAS:**

+ Para que o cĂłdigo abaixo funcione corretamente, o nome do seu classificador treinado anteriromente deve ser `model`.

**ObservaĂ§ĂŁo importante**

+ Como usamos simulaĂ§ĂŁo de monte carlo para calcular a SER do modelo de classificaĂ§ĂŁo, o tempo de execuĂ§ĂŁo da cĂ©lula abaixo pode ser longo.

````python
#=========SIMULAĂĂO MONTE CARLO - SER vs Eb/N0 para 8PSK com Gray Coding========
#=====Gray Coding â menor taxa de erro de bits (BER) para a mesma SER===========
# Detector: Classificador ML treinado (model.predict)
# =============================================

# =============================================
# CONFIGURAĂĂES:
# =============================================

N = 50_000_000                     # NĂșmero de sĂ­mbolos Monte Carlo

# Sweep em Eb/N0 com grid fino em alta SNR
EbN0dB = np.concatenate([
    np.arange(-2, 10, 2),
    np.arange(10, 14, 0.5),
])

k = 3                             # bits por sĂ­mbolo (8-PSK)
EsN0dB = EbN0dB + 10 * np.log10(k)

ser_simu = np.zeros(len(EbN0dB))
ser_theo = np.zeros(len(EbN0dB))

# =============================================
# CONSTELAĂĂO 8PSK COM GRAY CODING:
# =============================================

# Ordem Gray para 8-PSK (padrĂŁo usado em comunicaĂ§Ă”es)
gray_angles_deg = np.array([0, 45, 90, 135, 180, 225, 270, 315])

def modulate8PSK_gray(symbol_idx):
    """ModulaĂ§ĂŁo 8PSK com Gray Coding (energia normalizada Es = 1)"""
    theta = gray_angles_deg[symbol_idx] * np.pi / 180.0
    return np.exp(1j * theta)   # |s| = 1

# Precomputa a constelaĂ§ĂŁo Gray (usada apenas para geraĂ§ĂŁo dos sĂ­mbolos)
const_gray = np.array([modulate8PSK_gray(i) for i in range(8)])

# =============================================
# FUNĂĂES AUXILIARES:
# =============================================

def Q(x):
    return 0.5 * erfc(x / np.sqrt(2))

# =============================================
# SIMULAĂĂO:
# =============================================

print("Iniciando simulaĂ§ĂŁo 8-PSK com Gray Coding + Detector ML (model.predict)...\n")
print(f"{'Eb/N0(dB)':>8}   {'SER Simulado':>12}   {'SER TeĂłrico':>12}   {'Tempo (s)':>8}")
print("-" * 65)

start_total = time.time()

for idx in range(len(EbN0dB)):
    t0 = time.time()

    EsN0Lin = 10.0 ** (EsN0dB[idx] / 10.0)
    N0 = 1.0 / EsN0Lin

    # GeraĂ§ĂŁo de sĂ­mbolos (Ă­ndices 0-7 na ordem Gray)
    symbol_idx = np.random.randint(0, 8, N)
    symbol = modulate8PSK_gray(symbol_idx)

    # Canal AWGN
    noise = np.sqrt(N0 / 2.0) * (np.random.randn(N) + 1j * np.random.randn(N))
    y = symbol + noise

    # Deletando arrays que nĂŁo sĂŁo mais necessĂĄrias
    del symbol, noise

    # ====================== DETECTOR USANDO CLASSIFICADOR ML ======================
    # Converte sĂ­mbolo recebido para formato que o modelo espera (real + imag)
    X = np.c_[np.real(y), np.imag(y)]

    # Deletando arrays que nĂŁo sĂŁo mais necessĂĄrias
    del y

    # PrediĂ§ĂŁo com o modelo treinado anteriormente (Softmax Regression, etc.)
    detected_symbol = model.predict(X)

    # SER simulada
    errors = np.sum(detected_symbol != symbol_idx)
    ser_simu[idx] = errors / N

    # SER teĂłrica
    arg = np.sqrt(2 * EsN0Lin) * np.sin(np.pi / 8)
    ser_theo[idx] = 2 * Q(arg)

    elapsed = time.time() - t0
    print(f"{EbN0dB[idx]:8.1f}   {ser_simu[idx]:.3e}      {ser_theo[idx]:.3e}      {elapsed:7.2f}s")
print(f"\nTempo total: {time.time() - start_total:.1f} segundos")
print(f"\n SimulaĂ§ĂŁo concluĂ­da em {time.time() - start_total:.1f} segundos!")

# =============================================
# PLOT:
# =============================================
plt.figure(figsize=(10, 6))
plt.semilogy(EbN0dB, ser_theo, 'b-', linewidth=2.5, label='8PSK TeĂłrica')
plt.semilogy(EbN0dB, ser_simu, 'ro', markersize=6, label='8PSK Simulada (Gray + ML Classifier)')
plt.xlabel('Eb/N0 [dB]')
plt.ylabel('Symbol Error Rate (SER)')
plt.title('8PSK com Gray Coding + Detector ML treinado\n(SER vs Eb/N0 em canal AWGN)', fontsize=14, pad=15)
plt.grid(True, which='both')
plt.legend(fontsize=12)
plt.xlim([-2, 14])
plt.ylim([9e-6, 1])
plt.show()
````

Portanto, neste item, faremos uma busca por um limiar de classificaĂ§ĂŁo que melhore o desempenho global do classificador.

Para verificar qual Ă© o melhor limiar de classificaĂ§ĂŁo, o cĂłdigo abaixo faz:
  1. Gera probabilidades com o conjunto de validaĂ§ĂŁo (predict_proba)
  2. Testa vĂĄrios limiares de classificaĂ§ĂŁo no intervalo entre 0 e 1
  3. Escolhe o limiar que maximiza o F1-score, pois Ă© a mĂ©trica que garante equilĂ­brio entre falsos positivos e falsos negativos.

+ Aqui, nĂłs usamos o modelo treinado com o parĂąmetro `class_weight='balanced'`.
